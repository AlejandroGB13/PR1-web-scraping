```python
import pandas as pd
import os
```

```python
small_dataframes = []
csv_files = os.listdir('./data/')
for small_df in csv_files:
    df = pd.read_csv('./data/' + small_df, sep = ';', header = 0)
    small_dataframes.append(df)
df = pd.concat(small_dataframes).drop_duplicates().reset_index(drop=True)
df = df.sort_values('Name')
df = df.reset_index()
df.drop(columns=df.columns[0], axis=1, inplace=True)
df.drop(columns=df.columns[0], axis=1, inplace=True)
df
```

```python
df.to_csv('./coins_dataset.csv', sep=';')
```
